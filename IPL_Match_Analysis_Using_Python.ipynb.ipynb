## Data Cleaning :

```python
#check null values in matches data sets
matches.isnull().sum()
```

```python
# Check null values in deliveries dataset
deliveries.isnull().sum()
```

```python
#check missing value percentage in matches dataset
matches.isnull().sum()/len(matches)*100 
```

```python
#check missing value percentage in deliveries dataset
deliveries.isnull().sum()/len(deliveries)*100
```

```python
# Drop irrelevant Columns in matches dataset
matches.drop(columns=["umpire1","umpire2","umpire3"], inplace=True)
```

```python
# Handling null values in the Matches dataset
matches['city'].fillna('Unknown', inplace=True)
matches['winner'].fillna('No Result', inplace=True)
matches['player_of_match'].fillna('No Award', inplace=True)

# Verify if all null values are handled
print(matches.isnull().sum())
```

```python
#Drop irrelevant Columns in deliveries dataset
deliveries.drop(columns=["fielder"],inplace=True)
```

```python
# Handling Null Values in deliveries dataset
deliveries['player_dismissed'] = deliveries['player_dismissed'].fillna('Unknown')
deliveries['dismissal_kind'] = deliveries['dismissal_kind'].fillna('Not Applicable')
# Verify its all null value are Verified 
print(deliveries.isnull().sum())
```

```python
# Check Duplicated Values in Matches dataset
matches.duplicated().sum() #Counts the total number of duplicate rows in the DataFrame.
```

```python
# Check Duplicated Values in deliveries dataset
deliveries.duplicated().sum() # Counts the total number of duplicate rows in the DataFrame.
```

```python
# Count Duplicate Value in Each Row
deliveries.apply(lambda x:x.duplicated().sum())
```

```python
# Drop duplicates Values in deliveries dataset
deliveries.drop_duplicates(inplace=True)

# Check its all duplicated values are  Drop
print(deliveries.isnull().sum())
```

```python
# Correct Some Columns Name in Mates dataset
matches.rename(columns={"team1": "team_1", "team2": "team_2", "player_of_match":"player_of_the_match"}, inplace=True)
matches.head()
```

# Exploratory Data Analysis (EDA) :

Match-Level Analysis:


Q-1: Which team won the most matches:

```python
most_winning_team = matches['winner'].value_counts().idxmax()  # Team name
most_wins_count = matches['winner'].value_counts().max()       # Number of wins


print(f"The team with the most wins is {most_winning_team} with {most_wins_count} wins.")
```

Q-3: Average winning margin (runs aur wickets)?

```python

average_win_runs=matches["win_by_runs"].mean()
print("Average win by runs:",average_win_runs)

average_win_wicket=matches["win_by_wickets"].mean()
print("Average win by wickets:",average_win_wicket)
print()
```

Q-3: Top 5 cities  where matches were held:

```python
print("Top 5 cities where matches were held:")
top_cities = matches['city'].value_counts().head(5)
for city, count in top_cities.items():
   print(f"   - {city}: {count} matches")
```

Q-4: Venue with most match hosted:

```python
top_venue = matches['venue'].value_counts().idxmax()
top_venue_count=matches["venue"].value_counts().max()
print(f"Top venue with most Matches Hosted {top_venue} with {top_venue_count} matches")
print()
```

Q-5: Player with the most "Player of the match" Award:

```python
top_player = matches['player_of_the_match'].value_counts().idxmax()
player_awards = matches['player_of_the_match'].value_counts().max()
print(f"The player with the most 'Player of the Match' awards is {top_player} with {player_awards} awards.")
print()
```

Ball-Level Analysis:

Q-6: Batsman scored most runs overall:

```python
batsman_name = deliveries.groupby('batsman')['batsman_runs'].sum().idxmax()
batsman_runs = deliveries.groupby('batsman')['batsman_runs'].sum().max()
print(f"The batsman with the most runs is {batsman_name} with {batsman_runs} runs.")
print()
```

Q-7: Bowler with the most wickets:

```python
# Filter rows where a player was dismissed and dismissal_kind is valid
wickets_data = deliveries[(~deliveries['player_dismissed'].isna()) & (~deliveries['dismissal_kind'].isna())]

#  Count dismissals for each bowler
wickets_count = wickets_data['bowler'].value_counts()

# Find the bowler with the most wickets
most_wickets = wickets_count.idxmax()
wickets_taken = wickets_count.max()

print(f"The bowler with the most wickets is :- \n{most_wickets} with {wickets_taken} wickets.")
print()
```

Q-8: Distribution of extras:

```python
extra_column=["wide_runs","noball_runs","bye_runs","legbye_runs"] 
total_distributions=deliveries[extra_column].sum()
print("Distribution of extras:-","\n",total_distributions)
print()
```

Q-9: Team scored highest runs in a single match:

```python
# Group data by match_id and batting_team, then calculate total runs
team_scores = deliveries.groupby(['match_id', 'batting_team'])['total_runs'].sum().reset_index()

# Find the row with the highest total runs
highest_score = team_scores.loc[team_scores['total_runs'].idxmax()]

# Print the result
print("Match ID:", highest_score['match_id'])
print("Team:", highest_score['batting_team'])
print("Highest Runs Scored:", highest_score['total_runs'])
```

Q-10: Trend of total runs scored per over in a match:

```python
run_per_over=deliveries.groupby("over")["total_runs"].sum().reset_index()
run_per_over.plot(x="over",y="total_runs",kind="line",marker="*",color="green")
plt.xlabel("Over")
plt.ylabel("Total_Runs")
plt.title("Total Runs in Per Over")
plt.grid(True)
plt.savefig("Total Runs in Per Over.png")
plt.show()
```

Data Vizulazation:

```python
# Top 5 teams with the most wins
top_teams = matches['winner'].value_counts().head(5)

# Bar Chart
plt.figure(figsize=(8,5))
plt.bar(top_teams.index, top_teams.values, color='skyblue')
plt.title('Top 5 Teams with Most Wins')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.savefig('Top 5 Teams with Most Wins.png')  # Save chart as PNG
plt.show()
```

```python
# Top 5 batsmen with the highest runs
batsman_runs = deliveries.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(5)

# Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(batsman_runs.index, batsman_runs.values, color='orange')
plt.title('Top 5 Batsmen with the Highest Runs')
plt.xlabel('Batsmen')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.savefig('Top_5_Batsmen_with_Highest Runs.png')  # Save chart as PNG
plt.show()
```

```python
# Winning Margin Distribution - Runs and Wickets
plt.figure(figsize=(12, 6))

# Winning by runs
plt.subplot(1, 2, 1)
plt.hist(matches['win_by_runs'],color='blue')
plt.title('Winning Margin (Runs) Distribution')
plt.xlabel('Runs')
plt.ylabel('Frequency')

# Winning by wickets
plt.subplot(1, 2, 2)
plt.hist(matches['win_by_wickets'],color='green')
plt.title('Winning Margin (Wickets) Distribution')
plt.xlabel('Wickets')
plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('Winning Margin (Wickets) Distribution.png')  # Save chart as PN
plt.show()
```

```python
#Line plot showing runs scored across overs in a specific match
# Select a specific match_id (e.g., 1)
match_id = 1
match_over_runs = deliveries[deliveries['match_id'] == match_id].groupby('over')['total_runs'].sum()

# Line Plot
plt.figure(figsize=(10, 5))
plt.plot(match_over_runs.index, match_over_runs.values, marker='o', linestyle='-', color='purple')
plt.title(f'Total Runs Scored per Over - Match {match_id}')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.grid(True)
plt.savefig("Total Runs Scored per Over- Match.png")
plt.show()
```
